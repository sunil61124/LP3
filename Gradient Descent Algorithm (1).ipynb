Implement Gradient Descent Algorithm to find the local minima of a function.
For example, find the local minima of the function y=(x+3)Â² starting from the point x=2.

```python
# Import necessary libraries
import numpy as np
```

```python
# Define the function and its derivative
def function(x):
    """The function we want to minimize: y = (x + 3)^2"""
    return (x + 3) ** 2
```

```python
def derivative(x):
    """The derivative of the function: dy/dx = 2 * (x + 3)"""
    return 2 * (x + 3)
```

```python
# Parameters
x = 2                   # Starting point
learning_rate = 0.1     # Learning rate
iterations = 100        # Maximum number of iterations
tolerance = 1e-6        # Tolerance for stopping criterion
```

```python
# List to store values of x for each iteration (for visualization)
x_values = [x]
```

```python
# Gradient Descent Loop
for i in range(iterations):
    grad = derivative(x)            # Compute the gradient
    new_x = x - learning_rate * grad  # Update x by moving against the gradient

    # Store the new x value
    x_values.append(new_x)

    # Check if the change is smaller than the tolerance
    if abs(new_x - x) < tolerance:
        print(f"Converged after {i+1} iterations.")
        break

    x = new_x  # Update x for the next iteration
```

```python
# Results
print("The local minimum occurs at x =", x)
print("Minimum value of function is y =", function(x))
```

```python
# Optional: Plotting the function and the path of x values
import matplotlib.pyplot as plt

# Generate x values for plotting the function
x_plot = np.linspace(-10, 5, 100)
y_plot = function(x_plot)

# Plot the function
plt.plot(x_plot, y_plot, label="y = (x + 3)^2")
plt.scatter(x_values, [function(x) for x in x_values], color='red', label="Gradient Descent Path")
plt.title("Gradient Descent to Find Local Minimum")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

Explanation:
Function and Derivative: We define the target function 
y=(x+3)^2 and its derivative 2(x+3).
Parameters: x=2, a learning rate, maximum iterations, and a tolerance for stopping the loop.
Gradient Descent Loop: We iteratively update x by moving against the gradient and stop if the change in
x is smaller than the tolerance.
Visualization: We use matplotlib to plot the function and the path that Gradient Descent takes to find the local minimum.
